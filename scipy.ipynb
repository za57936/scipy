```python
import numpy as np      #importing numpy library
temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])     #storing given temperature data
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
from scipy import optimize        #importing optimize package from scipy package
def yearly_temps(times, avg, ampl, time_offset):     #fitting data array to the periodic function
    return (avg+ ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))
months = np.arange(12)
res_max, cov_max = optimize.curve_fit(yearly_temps, months,temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,temp_min, [-40, 20, 0])

import matplotlib.pyplot as plt       #importing matplotlib packages
%matplotlib inline


days = np.linspace(0, 12, num=365)      #plotting graph by setting appropriate values
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')        #setting lables to the plot
plt.ylabel('Min and max temperature')
```

```python
import pandas as pd     #importing pandas library
titanic = pd.read_csv("titanic.csv")      #loading titanic.csv file
titanic.head()
```

```python
males = titanic[titanic['sex']=='male'].index.value_counts().count()    #counting number of males
females = titanic[titanic['sex']=='female'].index.value_counts().count()    #counting number of females
lst = [males,females]    #storing number of males and females in list 'lst'
gender = ['male','female']
colors = ['r', 'g']

plt.pie(lst, labels=gender, colors=colors, startangle=90, autopct='%.1f%%')   #1-plotting pie chart
plt.show()
```

```python
#2-scatterplot with the Fare paid and the Age, differ the plot color by gender

male = titanic[titanic['sex']=='male']        #collecting male and female dataset
female = titanic[titanic['sex']=='female']
fig = plt.figure(figsize=(18,5))       #setting canvas size

plt.plot(male.fare, male.age, '.b', label='Male')     #plotting the scattered graph
plt.plot(female.fare, female.age, '.r', label='Female')
plt.axis('equal')
plt.xlabel('Fare')
plt.ylabel('Age')

leg = plt.legend()    #defining the labels
```
